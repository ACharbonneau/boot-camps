```python
# Load and sort data
stock = mlab.csv2rec(fname)
stock.sort(order='date')
```

```python
# Let's look at the first few records
stock[:5]
```

```python
# Pick only data since 1990 for clarity in figure (otherwise it gets crowded)
import datetime
s = stock[stock.date > datetime.date(1990, 1, 1)]
```

```python
# Simple trading volume analysis
dv = s.volume*s.close
print 'Min volume: $ %6.3f Million, on %s' %      (dv.min()/1e6, s.date[dv.argmin()])
print 'Max volume: $ %6d Million, on %s' % (int(dv.max()/1e6),s.date[dv.argmax()])
```

```python
plot(dv)
title('Daily volume')
figure()
plot(s.adj_close)
title('Adjusted closing price')
```

Let's find which days had a higher trading volume than a certain threshold (as a fraction):

```python
threshold = 0.5 # a number in the 0-1 range
mask = dv > threshold*dv.max()
# Let's print these dates
s.date[mask]
```

For better insight, let's plot trading volume and price, highlighting these high-volume days

```python
# We make a figure a bit larger than our defaults and with a stack of 2 plots (2 rows X 1 column):
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,6))

# On the top panel, plot the stock price as a function of date
ax1.plot(s.date, s.adj_close)
# Highlight the points above our threshold
ax1.scatter(s.date[mask], s.adj_close[mask], s=40, color='r', zorder=10)
ax1.set_ylabel('Stock Price ($)')
ax1.grid(True)

# On the bottom panel, plot the Trading volume also as a function of date
dvb = dv/1e9 # in billions
ax2.plot(s.date, dvb)
ax2.scatter(s.date[mask], dvb[mask], s=40, color='r',zorder=10)
ax2.set_ylabel('Volume ($ billion)')
ax2.grid(True)
```

Let's now fine-tune the figure a bit to get a polished final result

```python
# Hide ax1 x ticks
ax1.set_xticklabels(ax1.get_xticklabels(), visible=False)
# orient dates nicely
fig.autofmt_xdate()
# Make subplots close to each other
fig.subplots_adjust(hspace=0)
fig.suptitle('Ticker: %s' % ticker)
fig
```
