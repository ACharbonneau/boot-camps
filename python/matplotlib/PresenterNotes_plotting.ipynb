#Plotting in python

Plotting in python is very powerful. This is one of the reasons I decided to learn python, especially the interactive plotting.

##Preliminaries:
###pylab vs pyplot
pyplot is a module within matplotlib. pylab is a module which includes pyplot, numpy and some other packages. pylab will automatically give you the functionality of all of its included modules. For clarity, we will explicitly call pyplot and numpy separately.
###inline plotting
The ipython notebook can either pull up an interactive window or it can plot inline to the ipython notebook (as you see in this notebook). To get it to plot in the notebook you need to start the ipython notebook with the following command

<code> ipython notebook --pylab inline </code>

###Interactive plotting
In general, the default is for plotting to be interactive. This means that python will pop up a window and show you what you are plotting. There are times when you are plotting 100 plots, where you don't want this behavior. Therefore you can turn off interactive plotting by typing (in a python interpreter):

<code>pyplot.ioff() </code>

Alternately, if you try to plot something and nothing happens, make sure that interactive plotting is turned on:

<code> pyplot.ion() </code>


```python
#import numpy and pyplot
import numpy as np
from matplotlib import pyplot
```

```python
#read in data
x, y, err = np.genfromtxt('line_data.dat', unpack = True)
```

```python
#fit a line
coeff = np.polyfit(x, y, 1)
fit_y = np.polyval(coeff, x)
```

```python
#Plot the fit
pyplot.plot(x, fit_y)
#Show how awesome interactive plottin is...
```

###Line options

```python
#linestyle (ls)
pyplot.plot(x, fit_y, linestyle = ':')
```

```python
pyplot.plot(x, fit_y, linestyle = '--')
```

```python
#Linewidth (lw)
pyplot.plot(x, fit_y, linewidth = 3)
```

```python
#Color (c)
pyplot.plot(x, fit_y, color = 'r')
```

```python
#You can specify linestyle and color in a single string without a keyword
pyplot.plot(x, fit_y, '--g')
```

```python
#Add data
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, marker = '^')
```

###Symbol options

```python
#Remove connection from points:
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, marker = '^', linestyle = 'none')
```

*Note* that 'none' is a string, not the reserved work None. 

```python
#Alternative: you don't need to use the marker keyword
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, '^')
```

```python
#specify marker color
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, '^', markerfacecolor = 'c')
```

```python
#Alternately:
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, 'c^')
```

###axis limits and labels

```python
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, 'o')
#set x axis to run from 5 to 10
pyplot.xlim(5, 10)
```

```python
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, 'o')
pyplot.xlim(5, 10)
#set y axis to run from 5 to 10
pyplot.ylim(5, 10)
```

```python
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, 'o')
pyplot.xlim(5, 10)
pyplot.ylim(5, 10)

#Label axes
pyplot.xlabel('Days')
pyplot.ylabel('Salt concentration')
```

```python
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, 'o')
pyplot.xlim(5, 10)
pyplot.ylim(5, 10)
pyplot.xlabel('Days')
pyplot.ylabel('Salt concentration')
#Title plot:
pyplot.title('Salt Concentration as a function of days in the Sun')
```

###Add a legend

```python
pyplot.plot(x, fit_y, 'r')
pyplot.plot(x, y, 'o')
pyplot.xlim(5, 10)
pyplot.ylim(5, 10)
pyplot.xlabel('Days')
pyplot.ylabel('Salt concentration')
pyplot.title('Salt Concentration as a function of days in the Sun')
pyplot.legend(['fit', 'data'], loc = 'best')
```

```python
#Create a figure object
fig = pyplot.figure(1)
```

```python
fig.
```

```python
#create subplot object on that figure
ax = fig.add_subplot(1, 1, 1)  #rows, columns, plot number
```

```python
#You may need to use the pyplot.draw() command to see changes you make. If you expect something to change and it doesn't, try typing pyplot.draw()
```

```python
#plot onto the subplot object
ax.plot(x, fit_y, 'r')
ax.plot(x, y, 'o')
```

```python
# Now you have to set features rather than just invoking them (e.g. set_xlabel, set_xlim)
ax.plot(x, fit_y, 'r')
ax.plot(x, y, 'o')
ax.set_xlim(5, 10)
ax.set_ylim(5, 10)
ax.set_xlabel('Days')
ax.set_ylabel('Salt concentration')
ax.set_title('Salt Concentration as a function of days in the Sun')
```
